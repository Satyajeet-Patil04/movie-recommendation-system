```python
def cosine_similarity(a,b):
    dot=np.dot(a,b)
    moda=np.linalg.norm(a)
    modb=np.linalg.norm(b)
    return dot/(moda*modb)
```

```python
def s(a):
    for i in range(0,len(a)):
        for k in range(0,len(a)-1):
            if a[k+1]<a[k]:
                a[k+1],a[k]=a[k],a[k+1]
    return a
```
